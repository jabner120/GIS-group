### Purpose
This notebook first provides the LA city neighborhood median for the following variables: percentage of commute time above 30 minutes, percentage of commute time above 1 hour, median household income, percentage of vehicle ownership, and percentage of residents with a bachelor's degree or higher. The notebook then investigates the correlation between the dependent variable, commute time above 1 hour, and independent variables, household income, vehicle ownership, and bachelor's degree attainment or higher.
#### Data Source: 
[USC Neighborhood Data for Social Change - LA County Median Gross Rent](https://map.myneighborhooddata.org/?pa=42)

```python
import pandas as pd

# load la city data file
la = pd.read_csv (
    'data/LA_commute_educ.csv')
la.info()
```

```python
#Set values as string and remove '%'
la['Bachelors Degree or Higher'] = la['Bachelors Degree or Higher'].str.replace('%', '').astype(float)
la['Commute Time Above 1 Hour'] = la['Commute Time Above 1 Hour'].str.replace('%', '').astype(float)
la['Commute Time Above 30 Minutes'] = la['Commute Time Above 30 Minutes'].str.replace('%', '').astype(float)
la['Median Household Income'] = la['Median Household Income'].astype(float)
la['Vehicle Ownership'] = la['Vehicle Ownership'].str.replace('%', '').astype(float)
```

### Descriptive Stats on Commute Time in LA

```python
medians_to_find = ['Commute Time Above 1 Hour', 
                   'Median Household Income', 
                   'Vehicle Ownership', 
                   'Bachelors Degree or Higher']

medians = la[medians_to_find].median()
print(medians)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Create a box plot
sns.boxplot(y='Commute Time Above 1 Hour', data=la)

# Add titles and labels (optional)
plt.ylabel('Commute Time Above 1 Hour (%)')

# Show the plot
plt.show()
```

### Plot relationship between median household income and commute time

```python
import seaborn as sns

sns.set_theme(style="darkgrid")

# Create a jointplot
g = sns.jointplot(
    x="Median Household Income",
    y="Commute Time Above 1 Hour",
    data=la,
    kind="reg",
    truncate=False
)

# Set the title of the jointplot
g.fig.suptitle('Relationship Between Household Income and Commute Time', fontsize=16)

# Adjust the title position to account for the default Seaborn title padding
g.fig.subplots_adjust(top=0.90)

# Set the axis titles
g.set_axis_labels('Median Household Income ($)', 'Percentage with Commute Time Above 1 Hour')

# Show the plot
sns.plt.show()
```

### Plot relationship between vehicle ownership and commute time

```python
import seaborn as sns

sns.set_theme(style="darkgrid")

# Create a jointplot
g = sns.jointplot(
    x="Vehicle Ownership",
    y="Commute Time Above 1 Hour",
    data=la,
    kind="reg",
    truncate=False
)

# Set the title of the jointplot
g.fig.suptitle('Relationship Between Vehicle Ownership and Commute Time', fontsize=16)

# Adjust the title position to account for the default Seaborn title padding
g.fig.subplots_adjust(top=0.90)

# Set the axis titles
g.set_axis_labels('Percentage of Vehicle Ownership', 'Percentage with Commute Time Above 1 Hour')

# Show the plot
sns.plt.show()
```

### Plot relationship between bachelor degree attainment and commute time

```python
import seaborn as sns

sns.set_theme(style="darkgrid")

# Create a jointplot
g = sns.jointplot(
    x="Bachelors Degree or Higher",
    y="Commute Time Above 1 Hour",
    data=la,
    kind="reg",
    truncate=False
)

# Set the title of the jointplot
g.fig.suptitle('Relationship Between Education and Commute Time', fontsize=16)

# Adjust the title position to account for the default Seaborn title padding
g.fig.subplots_adjust(top=0.90)

# Set the axis titles
g.set_axis_labels('Percentage with Bachelor’s Degree or Higher', 'Percentage with Commute Time Above 1 Hour')

# Show the plot
sns.plt.show()
```
